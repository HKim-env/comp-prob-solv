```python
import matplotlib.pyplot as plt
from ce_thermo import thermodynamic_properties_isolated
from ce_thermo import thermodynamic_properties_soc
from ce_thermo import thermodynamic_properties_soc_cfs


# Define the temperature range
T_values = np.linspace(300, 2000, 100)

# Arrays to store the thermodynamic properties
U_iso_values, F_iso_values, S_iso_values = [], [], []
U_soc_values, F_soc_values, S_soc_values = [], [], []
U_soc_cfs_values, F_soc_cfs_values, S_soc_cfs_values = [], [], []

# Compute thermodynamic properties for each temperature
for T in T_values:
    U_iso, F_iso, S_iso = thermodynamic_properties_isolated(T)
    U_soc, F_soc, S_soc = thermodynamic_properties_soc(T)
    U_soc_cfs, F_soc_cfs, S_soc_cfs = thermodynamic_properties_soc_cfs(T)
    
    # Store the results
    U_iso_values.append(U_iso)
    F_iso_values.append(F_iso)
    S_iso_values.append(S_iso)
    
    U_soc_values.append(U_soc)
    F_soc_values.append(F_soc)
    S_soc_values.append(S_soc)
    
    U_soc_cfs_values.append(U_soc_cfs)
    F_soc_cfs_values.append(F_soc_cfs)
    S_soc_cfs_values.append(S_soc_cfs)

# Plot Internal Energy for the three cases
plt.figure(figsize=(10, 6))
plt.plot(T_values, U_iso_values, label='Isolated Ce3+')
plt.plot(T_values, U_soc_values, label='Ce3+ with SOC')
plt.plot(T_values, U_soc_cfs_values, label='Ce3+ with SOC & CFS')
plt.xlabel('Temperature (K)')
plt.ylabel('Internal Energy (J)')
plt.title('Internal Energy vs Temperature')
plt.legend()
plt.grid(True)
plt.show()

# Similarly, plots can be made for Free Energy and Entropy
```
